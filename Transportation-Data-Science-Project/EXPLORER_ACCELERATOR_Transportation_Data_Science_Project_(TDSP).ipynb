**TO DO:** Write 3-5 sentences about TSA.
> *  
Time Series Analysis (TSA) is the study of data points collected or recorded at specific time intervals to identify patterns and make predictions.
> *  A time series typically consists of three key components: trend, seasonality, and residual (irregular) components.
> *  The trend represents the long-term direction of the data, showing whether values are increasing, decreasing, or remaining stable.
> *  Seasonality refers to recurring patterns at fixed intervals, such as daily, monthly, or yearly fluctuations.
> *  The residual component captures random variations that cannot be explained by trend or seasonality.
> *  By decomposing a time series into these components using techniques like additive or multiplicative models, we can better understand its underlying structure and improve forecasting accuracy.






**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/Motor_Vehicle_Collisions_-_Crashes.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x=average_crashes_per_hour.index, y=average_crashes_per_hour.values)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Date')
plt.ylabel('Number of Crashes')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

Let's make a map that's inclusive for people with color blindness by also creating differently-shaped markers (squares, circles, and triangles) for crashes, injuries, and fatalities.

````python
#TODO: Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=42)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "Red"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "blue"  # Injuries
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
    else:
        color = "gray"  # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

**Identifying the Most Crash-Prone Intersections**

````python
import pandas as pd

# Filter out rows with missing intersection data
data_filtered = data.dropna(subset=['CROSS STREET NAME', 'ON STREET NAME' , 'CRASH DATE'])
````

````python
# Group by CROSS STREET NAME, BOROUGH, and CRASH DATE, then count crashes
intersection_crashes = data_filtered.groupby(['CROSS STREET NAME', 'BOROUGH', 'CRASH DATE']).size().reset_index(name='crash_count')

# Sort by the number of crashes in descending order
intersection_crashes_sorted = intersection_crashes.sort_values(by='crash_count', ascending=False)
````

**Top 10 Most Dangerous Intersections Visualization**

````python
import matplotlib.pyplot as plt

# Step 1: Strip whitespace from CROSS STREET NAME
intersection_crashes_sorted['CROSS STREET NAME'] = intersection_crashes_sorted['CROSS STREET NAME'].str.strip()

# Step 2: Group by CROSS STREET NAME and sum crash counts
grouped_df = intersection_crashes_sorted.groupby('CROSS STREET NAME', as_index=False)['crash_count'].sum()

# Step 3: Sort by crash_count in descending order
grouped_df = grouped_df.sort_values(by='crash_count', ascending=False)

# Step 4: Take the top 10 intersections
top_10_intersections = grouped_df.head(10)

# Step 5: Plot the grouped data
plt.figure(figsize=(10, 6))
plt.barh(top_10_intersections['CROSS STREET NAME'], top_10_intersections['crash_count'], color='red')
plt.xlabel('Number of Crashes')
plt.ylabel('Intersection')
plt.title('Top 10 Most Dangerous Intersections for Crashes not Fatality (Grouped by Name)')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest at the top
plt.show()
````

**Yearly Crash Trends at High-Risk Intersections so, Analyze how crash counts have changed over the years for the most dangerous intersections.**

````python
# Convert 'CRASH DATE' to datetime
intersection_crashes_sorted['CRASH DATE'] = pd.to_datetime(intersection_crashes_sorted['CRASH DATE'])

# Extract year from 'CRASH DATE'
intersection_crashes_sorted['YEAR'] = intersection_crashes_sorted['CRASH DATE'].dt.year

# Filter for top 10 intersections
top_intersections = top_10_intersections['CROSS STREET NAME'].tolist()
filtered_data = intersection_crashes_sorted[intersection_crashes_sorted['CROSS STREET NAME'].isin(top_intersections)]

# Group by year and intersection name, summing crashes
yearly_trends = filtered_data.groupby(['YEAR', 'CROSS STREET NAME'])['crash_count'].sum().unstack()

# Plot trend over years
plt.figure(figsize=(12, 6))
yearly_trends.plot(marker='o', colormap='tab10', figsize=(12, 6))
plt.title('Yearly Crash Trends at Top 10 Intersections')
plt.xlabel('Year')
plt.ylabel('Number of Crashes')
plt.legend(title='Intersection', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
````

**Day of the Week Analysis for Top Intersections Checking if certain days of the week see more crashes at specific intersections.
Crash Patterns by Day of the Week**

````python
# Extract day of the week from 'CRASH DATE'
intersection_crashes_sorted['DAY_OF_WEEK'] = intersection_crashes_sorted['CRASH DATE'].dt.day_name()

# Group by intersection and day of the week
day_trends = intersection_crashes_sorted.groupby(['DAY_OF_WEEK', 'CROSS STREET NAME'])['crash_count'].sum().unstack()

# Reorder days for better visualization
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_trends = day_trends.reindex(day_order)

# Filter for top intersections
day_trends_top = day_trends[top_intersections]

# Plot crash distribution by day
plt.figure(figsize=(12, 6))
day_trends_top.plot(kind='bar', colormap='plasma', figsize=(12, 6))
plt.title('Crash Counts by Day of the Week at Top 10 Intersections')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45)
plt.legend(title='Intersection', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

**Overall Crash Distribution by Day of the Week (2013-2024)**

````python
# Convert 'CRASH DATE' to datetime
intersection_crashes_sorted['CRASH DATE'] = pd.to_datetime(intersection_crashes_sorted['CRASH DATE'])

# Filter data for the years 2013-2024
intersection_crashes_sorted = intersection_crashes_sorted[
    (intersection_crashes_sorted['CRASH DATE'].dt.year >= 2013) &
    (intersection_crashes_sorted['CRASH DATE'].dt.year <= 2024)
]

# Extract day of the week from 'CRASH DATE'
intersection_crashes_sorted['DAY_OF_WEEK'] = intersection_crashes_sorted['CRASH DATE'].dt.day_name()

# Group by 'DAY_OF_WEEK' and sum the 'crash_count'
day_grouped = intersection_crashes_sorted.groupby('DAY_OF_WEEK')['crash_count'].sum()

# Sort the data by total crashes
day_grouped_sorted = day_grouped.sort_values(ascending=False)

# Plot the results
day_grouped_sorted.plot(kind='bar', figsize=(10, 6), colormap='viridis')

plt.title('Total Crashes by Day of the Week (2013-2024)')
plt.xlabel('Day of the Week')
plt.ylabel('Total Number of Crashes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print the day with the highest number of crashes
print(f"The day with the highest number of crashes is: {day_grouped_sorted.idxmax()} with {day_grouped_sorted.max()} crashes.")
````

**Monthly Crash Trends Over the Years (2013-2024)**

````python
# Convert 'CRASH DATE' to datetime
intersection_crashes_sorted['CRASH DATE'] = pd.to_datetime(intersection_crashes_sorted['CRASH DATE'])

# Extract the month name from 'CRASH DATE'
intersection_crashes_sorted['MONTH'] = intersection_crashes_sorted['CRASH DATE'].dt.month_name()

# Group by 'MONTH' and sum crash counts
month_grouped = intersection_crashes_sorted.groupby('MONTH')['crash_count'].sum()

# Order months chronologically
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

month_grouped = month_grouped.reindex(month_order)

# Plot the monthly trend
plt.figure(figsize=(12, 6))
month_grouped.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Total Crashes by Month (2013-2024)')
plt.xlabel('Month')
plt.ylabel('Total Number of Crashes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Print the month with the highest number of crashes
print(f"The month with the highest number of crashes is: {month_grouped.idxmax()} with {month_grouped.max()} crashes.")
````
